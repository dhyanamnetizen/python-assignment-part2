```python
# 1. Print menu in grouped format of categories
categories = ["Starters", "Mains", "Desserts"]

for cat in categories:
    print(f"----- {cat} -----")
    for item, details in menu.items():
        if details["category"] == cat:
            availability = "[Available]" if details["available"] else "[Unavailable]"
            print(f"{item:<15} ₹{details['price']:>7.2f}  {availability}")
    print()

# 2. Dictionary methods
total_items = len(menu)
available_count = sum(1 for item in menu.values() if item["available"])

# Searching the most expensive item
most_expensive_item = max(menu, key=lambda x: menu[x]["price"])
expensive_price = menu[most_expensive_item]["price"]

# Items with price less than 150
cheap_items = [name for name, info in menu.items() if info["price"] < 150]

print(f"Total items on menu: {total_items}")
print(f"Total available items: {available_count}")
print(f"Most expensive item: {most_expensive_item} (₹{expensive_price})")
print(f"Items priced under ₹150: {', '.join(cheap_items)}")
```

```python
cart = []

def add_to_cart(item_name, quantity):
    # Checking existence and availability
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in the menu.")
        return
    if not menu[item_name]["available"]:
        print(f"Error: '{item_name}' is currently unavailable.")
        return
    
    # Checking if the item is already in the cart, so that we can update the quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f"Updated {item_name} quantity to {entry['quantity']}.")
            return
            
    # If the item has not been added to the cart, then add to cart
    cart.append({"item": item_name, "quantity": quantity, "price": menu[item_name]["price"]})
    print(f"Added {quantity} x {item_name} to cart.")

def remove_from_cart(item_name):
    global cart
    initial_len = len(cart)
    cart = [entry for entry in cart if entry["item"] != item_name]
    if len(cart) < initial_len:
        print(f"Removed {item_name} from cart.")
    else:
        print(f"Message: {item_name} was not in the cart.")

# Sequence
print("--- Simulating Cart Operations ---")
add_to_cart("Paneer Tikka", 2)
add_to_cart("Gulab Jamun", 1)
add_to_cart("Paneer Tikka", 1) # Should update to 3
add_to_cart("Mystery Burger", 1) # Should show error
add_to_cart("Chicken Wings", 1) # Should show unavailable
remove_from_cart("Gulab Jamun")

# 5. Final Order Summary
print("\n--- Final Order Summary ---")
subtotal = 0
for entry in cart:
    item_total = entry["quantity"] * entry["price"]
    subtotal += item_total
    print(f"{entry['item']:15} x{entry['quantity']}   ₹{item_total:>7.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst

print("-" * 30)
print(f"{'Subtotal:':15} ₹{subtotal:>7.2f}")
print(f"{'GST (5%):':15} ₹{gst:>7.2f}")
print(f"{'Total Payable:':15} ₹{total_payable:>7.2f}")
```

```python
# 1 & 2. Printing revenue per day and best-selling day
daily_revenue = {}
for date, orders in sales_log.items():
    total_rev = sum(order["total"] for order in orders)
    daily_revenue[date] = total_rev

best_day = max(daily_revenue, key=daily_revenue.get)

print("Revenue per day:")
for date, rev in daily_revenue.items():
    print(f"{date}: ₹{rev:.2f}")
print(f"\nBest-selling day: {best_day} (₹{daily_revenue[best_day]:.2f})")

# 3. Most ordered item
item_counts = {}
for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            item_counts[item] = item_counts.get(item, 0) + 1

most_ordered = max(item_counts, key=item_counts.get)
print(f"Most ordered item across all days: {most_ordered}")

# 4. Adding of a new day
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 450.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

# 5. Printing a numbered list using enumerate function
print("\n--- Complete Order History ---")
all_orders = []
for date, orders in sales_log.items():
    for order in orders:
        all_orders.append((date, order))

for i, (date, order) in enumerate(all_orders, 1):
    items_str = ", ".join(order["items"])
    print(f"{i}. [{date}] Order #{order['order_id']} - ₹{order['total']:.2f} - Items: {items_str}")
```
